### 1.3 Download tweet data from API, save to `tweet_json.txt`, set a Datafram

```python
import tweepy
import json
```

```python
# Followed the request, I delete all true information.
consumer_key = 'my consumer key'
consumer_secret = 'my consumer secret'
access_token = 'my access token'
access_secret = 'my access secret'
```

```python
# scrap all tweet with the same tweet_id in `twitter_enhanced`
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
api = tweepy.API(auth, wait_on_rate_limit = True, wait_on_rate_limit_notify = True)
```

```python
n = 1
for tweid in twitter_enhanced.tweet_id:
    with open('tweet_json.txt','a' ) as file:
        try:
            tweet = api.get_status(tweid, tweet_mode='extended')
            json_tweet = json.dumps(tweet._json)
            file.write(json_tweet + '\n')
        except:
            file.write( 'Date has been deleted. \n')
    print('No.' , '{:>5}'.format(n) , tweid ,' finished')
    n += 1
```

I find if this tweet is a retweet, it will have a 'retweeted_status' sub_dict. I found the tweet with index 68 is a retweet, so I use json format to print a tweet and a retweet to find some details.

I copy the json formate text into Atom。

I find for tweet, all necessary contents are in the first level of the nested dicitonary. 
But if the tweet is a retweet, the 'favorite_count' and original ID is in the sub_dicitonary with keyname 'retweeted_status'

```python
# Load 'full_text', 'tweet_id', 'retweet_count', 'favorite_count', 
#'in_reply_to_status_id', 'retweeted_status_id' from 'tweet_json.txt', 
# and save them as a dicitonary in a list

df_list = [] #set a empty list

with open('tweet_json.txt','r') as jsfile:
    for line in jsfile:
        try:   
            jline = json.loads(line)  #load every row of the txt file in json mode
            full_text = jline["full_text"]
            tweet_id = jline["id_str"]
            in_reply_to_status_id = jline["in_reply_to_status_id_str"]
            
            if jline.get('retweeted_status') is None:  # if this tweet is not a retweet
                retweet_count = jline["retweet_count"]
                favorite_count = jline["favorite_count"]
                retweeted_status_id = None
            
            else:   # if this tweet is a retweet
                retweet_count = jline['retweeted_status']["retweet_count"]
                favorite_count = jline['retweeted_status']['favorite_count']
                retweeted_status_id = jline['retweeted_status']["id_str"]

            df_list.append({'tweet_id':tweet_id, 
                            'full_text': full_text, 
                            'retweet_count':retweet_count,
                            'favorite_count':favorite_count,
                            'in_reply_to_status_id':in_reply_to_status_id,
                            'retweeted_status_id' : retweeted_status_id})
        except:
            continue

#change df_list into a datafram called 'supplement'
supplement = pd.DataFrame(df_list)    
```

#### (13) Clean not dog predictions value of `image_predictions_clean` colmun

##### Define
some predictions are not breeds of dogs, such as ice_cream and tissue, need to clean these irrelevant predicitions

##### Code

```python
# replace non-dog prediction with next prediction, use np.nan to fill the gap
for i in range(image_predictions_clean.shape[0]):
    if image_predictions_clean.loc[i,'p3_dog'] == False:
        image_predictions_clean.loc[i,['p3','p3_conf','p3_dog']] = np.full(3,np.nan)
    if image_predictions_clean.loc[i,'p2_dog'] == False:
        image_predictions_clean.loc[i,['p2','p2_conf','p2_dog']] = \
                image_predictions_clean.loc[i,['p3','p3_conf','p3_dog']].to_numpy()
        image_predictions_clean.loc[i,['p3','p3_conf','p3_dog']] = np.full(3,np.nan)
    if image_predictions_clean.loc[i,'p1_dog'] == False:
        image_predictions_clean.loc[i,['p1','p1_conf','p1_dog']] = \
                image_predictions_clean.loc[i,['p2','p2_conf','p2_dog']].to_numpy()
        image_predictions_clean.loc[i,['p2','p2_conf','p2_dog']] = \
                image_predictions_clean.loc[i,['p3','p3_conf','p3_dog']].to_numpy()
        image_predictions_clean.loc[i,['p3','p3_conf','p3_dog']] = np.full(3,np.nan)

# delete rows with all prediction is null
image_predictions_clean = image_predictions_clean[image_predictions_clean.p1_dog.notnull()]

# now all content in the table are dog breed(all p1_dog,p2_dog,p3_dog columns have only True 
# value) I will drop these columns
image_predictions_clean = image_predictions_clean.drop(columns = {'p1_dog','p2_dog','p3_dog'})   
```
