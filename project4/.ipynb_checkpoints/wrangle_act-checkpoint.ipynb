### 1.3 Download tweet data from API, save to `tweet_json.txt`, set a Datafram

```python
import tweepy
import json
```

```python
# Followed the request, I delete all true information.
consumer_key = 'my consumer key'
consumer_secret = 'my consumer secret'
access_token = 'my access token'
access_secret = 'my access secret'
```

```python
# scrap all tweet with the same tweet_id in `twitter_enhanced`
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
api = tweepy.API(auth, wait_on_rate_limit = True, wait_on_rate_limit_notify = True)
```

```python
n = 1
for tweid in twitter_enhanced.tweet_id:
    with open('tweet_json.txt','a' ) as file:
        try:
            tweet = api.get_status(tweid, tweet_mode='extended')
            json_tweet = json.dumps(tweet._json)
            file.write(json_tweet + '\n')
        except:
            file.write( 'Date has been deleted. \n')
    print('No.' , '{:>5}'.format(n) , tweid ,' finished')
    n += 1
```

I have downloads all the possible files from API. I open this txt file with jupyter notebook, and use 'control+F' to look for columns I need.

I find if this tweet is a retweet, it will have a 'retweeted_status' sub_dict. I found the tweet with index 68 is a retweet, so I use json format to print a tweet and a retweet to find some details.

```python
#print a tweet and a retweet with json format
with open('tweet_json.txt','r') as jsfile:
    lines = jsfile.readlines()
    js_line1 = json.loads(lines[0])
    js_line2 = json.loads(lines[68])

print(json.dumps(js_line1,indent=4)) 
print('-----------------------------')
print(json.dumps(js_line2,indent=4)) 
```

I copy the json formate text into Atom and find most of the interesting contents are at the first level of the nested dicitonary. 
But if the tweet is a retweet, the 'favorite_count' and original ID is in the sub_dicitonary with keyname 'retweeted_status'

```python
# Load 'full_text', 'tweet_id', 'retweet_count', 'favorite_count', 
#'in_reply_to_status_id', 'retweeted_status_id' from 'tweet_json.txt', 
# and save them as a dicitonary in a list

df_list = [] #set a empty list

with open('tweet_json.txt','r') as jsfile:
    for line in jsfile:
        try:   
            jline = json.loads(line)  #load every row of the txt file in json mode
            full_text = jline["full_text"]
            tweet_id = jline["id_str"]
            in_reply_to_status_id = jline["in_reply_to_status_id_str"]
            
            if jline.get('retweeted_status') is None:  # if this tweet is not a retweet
                retweet_count = jline["retweet_count"]
                favorite_count = jline["favorite_count"]
                retweeted_status_id = None
            
            else:   # if this tweet is a retweet
                retweet_count = jline['retweeted_status']["retweet_count"]
                favorite_count = jline['retweeted_status']['favorite_count']
                retweeted_status_id = jline['retweeted_status']["id_str"]

            df_list.append({'tweet_id':tweet_id, 
                            'full_text': full_text, 
                            'retweet_count':retweet_count,
                            'favorite_count':favorite_count,
                            'in_reply_to_status_id':in_reply_to_status_id,
                            'retweeted_status_id' : retweeted_status_id})
        except:
            continue

#covert the list into a datafram 'supplement'
supplement = pd.DataFrame(df_list)    
```

#### 5.combine all 4 columns stages of `twitter_enhanced`table into 1 column

##### Define
- a. use pd.melt to set a new table, in that table combine 'pupper', 'doggo', 'puppo','floofer'columns into one column named 'stage',values are 'pupper', 'doggo', 'puppo','floofer' or 'None'
- b. drop all rows with value 'None'
- c. Then use a left join to combine twitter_enhanced_clean and the new table(choosen the necessary columns)

##### Code

```python
# a. use melt method to build up a new datafram 'df_stage' convert all 4 columns into one column
df_stage = pd.melt(twitter_enhanced_clean,id_vars=['tweet_id', 'in_reply_to_status_id', 'in_reply_to_user_id', 'timestamp',
       'source', 'text', 'retweeted_status_id', 'retweeted_status_user_id',
       'retweeted_status_timestamp', 'expanded_urls', 'rating_numerator',
       'rating_denominator', 'name','retweet_count', 'favorite_count'],value_name = 'stage')
# b. drop all values equal to 'None'
df_stage = df_stage[df_stage.stage != 'None']

# c. merge this small datafrom to twitter_enhanced_clean using leftjoin
twitter_enhanced_clean = pd.merge(twitter_enhanced_clean[['tweet_id', 'in_reply_to_status_id', 'in_reply_to_user_id', 'timestamp',
       'source', 'text', 'retweeted_status_id', 'retweeted_status_user_id',
       'retweeted_status_timestamp', 'expanded_urls', 'rating_numerator',
       'rating_denominator', 'name','retweet_count', 'favorite_count']],
             df_stage[['tweet_id','stage']],on = 'tweet_id',how = 'left')
             
```
